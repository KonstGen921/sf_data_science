```python
names = ['chlorhexidine', 'cyntomycin', 'afobazol'] 
counts = [15, 18, 7]
import pandas as pd
def create_medications(names, counts):
    medications= pd.Series(data = counts, index=names, name='medications')
    return medications
med_series = create_medications(names, counts)
def get_percent(med_series):
    total = med_series.sum()
    percent_series = (med_series/total)
    return(percent_series)
percent_series= get_percent(med_series)
print(med_series, percent_series, sep='\n')
```

```python
import pandas as pd
income = [478, 512, 196]
expenses = [156, 130, 270]
years = [2018, 2019, 2020]
def create_companyDF(income, expenses, years):
    data = {'income':income, 'expenses':expenses}
    companyDF = pd.DataFrame(data, index = years)
    companyDF['profit'] = companyDF['income'] - companyDF['expenses']
    return(companyDF)
companyDF = create_companyDF(income, expenses, years)

print(companyDF.loc[2019,'expenses'])
```

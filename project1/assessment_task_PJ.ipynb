Объединил таблицы. Убрал лишнюю колонку Unnamed: 0. Во второй таблице одни и теже анализы задвоены под разными названиями. Удаляем повторные столбцы.

```python
# будем избавляться от выбросов в age, bmi, hdl, sbp, fib4
def outliers_iqr(data, column):
    x = data[column]
    quartile_1 = x.quantile(0.25)
    quartile_3 = x.quantile(0.75)
    iqr = quartile_3 - quartile_1
    lower_bound = quartile_1 - 1.5 * iqr
    upper_bound = quartile_3 + 1.5 * iqr
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    return outliers, cleaned
ouliers_age, cleaned_age = outliers_iqr(merged_df, 'age')
display(ouliers_age.shape[0], cleaned_age.shape[0])
sns.boxplot(data=cleaned_age, x='age')
plt.title('Распределение по возрасту без выбросов')
plt.show()

ouliers_bmi, cleaned_bmi = outliers_iqr(merged_df, 'bmi')
display(ouliers_bmi.shape[0], cleaned_bmi.shape[0])
sns.boxplot(data=cleaned_bmi, x='bmi')
plt.title('Распределение индекса массы тела (BMI) без выбросов')
plt.show()

ouliers_hdl, cleaned_hdl = outliers_iqr(merged_df, 'hdl')
display(ouliers_hdl.shape[0], cleaned_hdl.shape[0]) 
sns.boxplot(data=cleaned_hdl, x='hdl')
plt.title('Распределение липопротеинов высокой плотности (HDL) без выбросов')
plt.show()

ouliers_sbp, cleaned_sbp = outliers_iqr(merged_df, 'sbp')
display(ouliers_sbp.shape[0], cleaned_sbp.shape[0])
sns.boxplot(data=cleaned_sbp, x='sbp')
plt.title('Распределение систолического давления (SBP) без выбросов')
plt.show()

outliers_fib4, cleaned_fib4 = outliers_iqr(merged_df, 'fib4')
display(outliers_fib4.shape[0], cleaned_fib4.shape[0])
sns.boxplot(data=cleaned_fib4, x='fib4')
plt.title('Распределение индекса фиброза печени (FIB-4) без выбросов')
plt.show()

Gender_counts = merged_df['male'].value_counts().reset_index(name='Count')
Gender_counts['male'] = Gender_counts['male'].map({0: 'Женщины' , 1: 'Мужчины'})
display(Gender_counts)
plt.pie(data=Gender_counts, x='Count', autopct='%1.1f%%', labels=['Женщины', 'Мужчины'])
plt.title('Распределение по полу')
plt.show()


death_counts = merged_df['status'].value_counts().reset_index(name='Count')
death_counts['status'] = death_counts['status'].map({0: 'Жив', 1: 'Умер'})
display(death_counts)
plt.pie(data=death_counts, x='Count', autopct='%1.1f%%', labels=['Жив', 'Умер'])
plt.title('Распределение по статусу')
plt.show()
```

```python
# Видно что здесь неадекватный возраст до 191 года поэтому сразу избавимся от выбросов
# Для начала сделаю маску что бы отбросить все что больше 90 лет
mask_nwtco_age = nwtco_df[nwtco_df['age'] < 100]

def outliers_iqr_nwtco(data, column):
    x = data[column]
    quartile_1 = x.quantile(0.25)
    quartile_3 = x.quantile(0.75)
    iqr = quartile_3 - quartile_1
    lower_bound = quartile_1 - 1.5 * iqr
    upper_bound = quartile_3 + 1.5 * iqr
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    return outliers, cleaned
outliers_nwtco_age, cleaned_nwtco_age = outliers_iqr_nwtco(mask_nwtco_age, 'age')
display(outliers_nwtco_age.shape[0], cleaned_nwtco_age.shape[0])
sns.boxplot(data=mask_nwtco_age, x='age')
plt.title('Распределение по возрасту в nwtco_df без выбросов')
plt.show()

# посмотрим на категориальные данные
instit_counts = mask_nwtco_age['instit'].value_counts().reset_index(name='count')
display(instit_counts)

sns.barplot(data=instit_counts, x='instit', y='count', hue='instit', palette='Set2')
plt.title('Распределение по институтам')
plt.xlabel('Институт')
plt.ylabel('Количество')
plt.show()

histol_count = mask_nwtco_age['histol'].value_counts().reset_index(name='count')
display(histol_count)
sns.barplot(data=histol_count, x='histol', y='count', hue='histol', palette='Set2')
plt.title('Распределение по гистологии')
plt.xlabel('Гистология')
plt.show()

stage_counts = mask_nwtco_age['stage'].value_counts().reset_index(name='count')
display(stage_counts)
sns.barplot(data=stage_counts, x='stage', y='count', hue='stage', palette='Set2')
plt.title('Распределение по стадиям')
plt.xlabel('Стадия')
plt.show()

study_counts = mask_nwtco_age['study'].value_counts().reset_index(name='count')
display(study_counts)
sns.barplot(data=study_counts, x='study', y='count', hue='study', palette='Set2')
plt.title('Распределение по исследованиям')
plt.xlabel('Исследование')
plt.show()

rel_counts = mask_nwtco_age['rel'].value_counts().reset_index(name='count')
display(rel_counts)
sns.barplot(data=rel_counts, x='rel', y='count', hue='rel', palette='Set2')
plt.title('Распределение по статусу рецидива')
plt.xlabel('Наличие рецидива')
plt.show()

# посмотрим на распределение по времени до рецидива

sns.boxplot(data=mask_nwtco_age, x='edrel')
plt.title('Распределение по времени до рецидива')
plt.xlabel('Время до рецидива (дни)')
plt.show()
```
